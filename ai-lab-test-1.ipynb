```python
def heuristic(state, target):
    d = 0
    for i in [-1,1,2,3,4]:
        d1, d2 = state.index(i), target.index(i)
        x1, y1 = d1 % 3, d1 // 3
        x2, y2 = d2 % 3, d2 // 3
        d += abs(x1-x2) + abs(y1-y2)
    return d
```

```python
def possible_moves(state, visited):
    b = state.index(-1)  
    d = []
    if 9 > b - 3 >= 0: 
        d += 'u'
    if 9 > b + 3 >= 0:
        d += 'd'
    if b not in [2,5,8]: 
        d += 'r'
    if b not in [0,3,6]: 
        d += 'l'
    moves = []
    for move in d:
        moves.append(gen(state,move,b))
    return [move for move in moves if tuple(move) not in visited]
```

```python
def gen(state, d, b):
    t = state.copy()
    if d == 'u':
        t[b-3], t[b] = t[b], t[b-3]
    if d == 'd':
        t[b+3], t[b] = t[b], t[b+3]
    if d == 'r':
        t[b+1], t[b] = t[b], t[b+1]
    if d == 'l':
        t[b-1], t[b] = t[b], t[b-1]
    return t
```

```python
def astar(src, target):
    states = [src]
    g = 0
    visited = set()
    while len(states):
        print("Level: ", g)
        moves = []
        for state in states:
            visited.add(tuple(state))
            print_grid(state)
            if state == target:
                print("Success.")
                return
            moves += [move for move in possible_moves(state, visited) if move not in moves]
        costs = [g + heuristic(move, target) for move in moves]
        g += 1
        states = [moves[i] for i in range(len(moves)) if costs[i] == min(costs)]
        
    print("Failed to find the goal.")
```

```python
#Auxiliary
def print_grid(state):
    t = state.copy()
    t[t.index(-1)] = ' '
    print(f"""\n{t[0]} {t[1]} {t[2]}
{t[3]} {t[4]} {t[5]}
{t[6]} {t[7]} {t[8]}\n""")
```

```python
src = ['*',2,4,'*',-1,'*','*',3,1]
target = [-1,1,2,3,4,'*','*','*','*']         

print_grid(src)
print_grid(target)

astar(src, target)
```
